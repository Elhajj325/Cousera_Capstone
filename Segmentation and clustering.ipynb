```python
import pandas as pd
import numpy as np
```

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
```

```python
url = "https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"
html = urlopen(url)
```

```python
soup = BeautifulSoup(html, 'lxml')
type(soup)
```

```python
 #Get the title
title = soup.title
print(title)
```

```python
# Print out the text
text = soup.get_text()
#print(soup.text)
```

```python
soup.find_all('a')
```

```python
all_links = soup.find_all("a")
for link in all_links:
    print(link.get("href"))
```

```python
all_tables=soup.find_all("table")
all_tables
```

```python
rows = [ row for row in soup.findAll('tr') if 'Not assigned' not in row.text ]
```

```python
right_table=soup.find('table', class_='wikitable sortable')
right_table
```

```python
A=[]
B=[]
C=[]

for row in right_table.findAll('tr'):
    cells=row.findAll('td')
    if len(cells)==3:
        A.append(cells[0].find(text=True))
        B.append(cells[1].find(text=True))
        C.append(cells[2].find(text=True))
```

```python
df=pd.DataFrame(A,columns=['Postal Code'])
df['Borough']=B
df['Neighborhood']=C
df
```

```python
df1 = df.replace('Not assigned',np.nan, regex=True)
df1.head()
```

```python
df1.dropna(subset=["Borough"], axis=0, inplace=True)
df1 = df1.reset_index(drop=True)
df1.head(12)
```

```python
df1.shape
```

```python
df2= pd.read_csv("https://cocl.us/Geospatial_data")
df2.head(10) 
```

```python
df3=df1.merge(df2, left_on ='Postal Code', right_on ='Postal Code')
df3.head()
```
